## üß† Step 5: Compute Final Label (Python Example)

```python
import numpy as np

def compute_health_label(row):
    def scale(value, bins, scores):
        # helper: assign score based on bins
        for i, b in enumerate(bins):
            if value < b:
                return scores[i]
        return scores[-1]

    liquidity = scale(row["Current_Ratio"], [0.8, 1.5, 2.0], [2, 5, 10])
    leverage = scale(row["Debt_to_Equity"], [0.5, 2, 3], [10, 5, 0])
    profitability = scale(row["ROE"], [0, 0.1, 0.2], [2, 5, 10])
    coverage = scale(row["Interest_Coverage"], [1, 3, 5], [2, 5, 10])
    cashflow = scale(row["Operating_CashFlow_to_Debt"], [0.1, 0.3, 0.5], [2, 5, 10])
    risk = scale(row["Net_FX_Position"], [-1e8, 0, 1e8], [0, 5, 10])
    
    # Weighted sum
    score = (0.2*liquidity + 0.2*leverage + 0.25*profitability + 
             0.2*cashflow + 0.1*coverage + 0.05*risk)
    return round(score, 1)
```

This outputs a **continuous score (0‚Äì10)** representing company health.
