```python
import pandas as pd
import numpy as np

# Colunas que identificam uma observação
id_vars = ['CompanyCode', 'Year', 'Description', 'OriginalFileName']

# Colunas que queremos transformar em linhas
value_vars = ['Annual', '9_Months', '6_Months', '3_Months']

# Aplicando a transformação "melt"
df_long = pd.melt(df,
                  id_vars=id_vars,
                  value_vars=value_vars,
                  var_name='Period', # Nova coluna para o período
                  value_name='Value') # Nova coluna para o valor

print("Formato original:")
print(df.head())
print("\nFormato 'long' após o melt:")
print(df_long.head())
```

```python
import pandas as pd

# Supondo que seu DataFrame original se chama 'df'
# df = pd.read_csv('seu_arquivo.csv')

# Definir as colunas que identificam um documento único
id_documento = ['CompanyCode', 'Year', 'OriginalFileName']

# Definir as colunas de valores onde os nulos podem ocorrer
colunas_valores = ['Annual', '9_Months', '6_Months', '3_Months']

# Agrupar por documento e agregar as informações
# Para cada documento, vamos contar:
# 1. O número total de métricas (linhas)
# 2. A soma de valores nulos em cada coluna de período
analise_nulos_por_documento = df.groupby(id_documento).agg(
    total_metricas=('Description', 'count'),
    nulos_annual=('Annual', lambda x: x.isnull().sum()),
    nulos_9m=('9_Months', lambda x: x.isnull().sum()),
    nulos_6m=('6_Months', lambda x: x.isnull().sum()),
    nulos_3m=('3_Months', lambda x: x.isnull().sum())
).reset_index()

# Calcular o total de nulos por documento
analise_nulos_por_documento['total_nulos'] = (
    analise_nulos_por_documento['nulos_annual'] +
    analise_nulos_por_documento['nulos_9m'] +
    analise_nulos_por_documento['nulos_6m'] +
    analise_nulos_por_documento['nulos_3m']
)

# Calcular a porcentagem de células de valor que são nulas em cada documento
# O total de células de valor é o número de métricas * 4 (as 4 colunas de período)
analise_nulos_por_documento['percentual_nulos'] = (
    analise_nulos_por_documento['total_nulos'] / (analise_nulos_por_documento['total_metricas'] * len(colunas_valores)) * 100
)

# Ordenar para ver os documentos com mais nulos no topo
documentos_mais_incompletos = analise_nulos_por_documento.sort_values(by='percentual_nulos', ascending=False)

# Exibir os 10 documentos mais incompletos
print("Top 10 documentos com maior percentual de valores nulos:")
print(documentos_mais_incompletos.head(10))
```
