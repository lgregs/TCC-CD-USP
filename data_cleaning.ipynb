```python
# Melt do df (versao alternativa --- analisar qual sera melhor para o AM)
id_vars = ['CompanyCode', 'Ano', 'Descricao', 'OriginalFileName']
value_vars = ['Anual', '9_Meses', '6_Meses', '3_Meses']

df_long = pd.melt(df,
                  id_vars=id_vars,
                  value_vars=value_vars,
                  var_name='Period', # Nova coluna para o período
                  value_name='Value') # Nova coluna para o valor

print("\nFormato 'long' após o melt:")
df_long.head()
```

```python
# Agrupar pela coluna 'Descricao'
analise_nulos_por_descricao = df.groupby('Descricao').agg(
    # Conta o total de vezes que a descrição aparece no dataset
    total_ocorrencias=('Descricao', 'count'),
    # Conta os nulos em cada coluna de período para essa descrição
    nulos_anual=('Anual', lambda x: x.isnull().sum()),
    nulos_9m=('9_Meses', lambda x: x.isnull().sum()),
    nulos_6m=('6_Meses', lambda x: x.isnull().sum()),
    nulos_3m=('3_Meses', lambda x: x.isnull().sum())
)

# Calcular o total de nulos para cada descrição
analise_nulos_por_descricao['total_nulos'] = (
    analise_nulos_por_descricao['nulos_anual'] +
    analise_nulos_por_descricao['nulos_9m'] +
    analise_nulos_por_descricao['nulos_6m'] +
    analise_nulos_por_descricao['nulos_3m']
)
```
