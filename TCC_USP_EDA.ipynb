```python
df = df.rename(columns={
    'Yıllık': 'Anual',
    '9 Aylık': '9 Meses',
    '6 Aylık': '6 Meses',
    '3 Aylık': '3 Meses'
})
print(df.columns)
```

```python
df.isnull().sum()
```

```python
valores_nulos_por_empresa = df.groupby('empresa').apply(lambda x: x.isnull().sum())
print(valores_nulos_por_empresa)
```

```python
df_bimas_sise = df[df['empresa'].isin(['BIMAS', 'SISE'])]
df_bimas_sise
```

```python
df_bimas_sise.isnull().sum()
```

```python
antigo = df_bimas_sise['ano'].min()
recente = df_bimas_sise['ano'].max()

print(f"Dados mais antigos: {antigo}")
print(f"Dados mais recentes: {recente}")
```
